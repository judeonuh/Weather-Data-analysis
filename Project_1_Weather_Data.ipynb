```python
# Filter the records where the weather was exactly "clear"
clear_weather = df[df['Weather'] == 'Clear']

# Display the filtered records
clear_weather
```

```python
# Find the number of times the wind speed was exactly 4 km/hr.
wind_speed_4 = df[df['Wind_Speed_km_h'] == 4]
num_wind_speed_4 = wind_speed_4.shape[0]
num_wind_speed_4
```

```python
# Renaming the column "Weather" to "Weather_Condition."
df_renamed = df.rename(columns={'Weather': 'Weather_Condition'})
df_renamed.head()
```

```python
# Find the mean visibility of the dataset
df_renamed.describe()
```

```python
# Filter the records where wind speed is greater than 24 km/h and visibility is equal to 25 km
wind_24_vis_25 = df_renamed[(df_renamed['Wind_Speed_km_h'] > 24) & (df['Visibility_km'] == 25)]

num_wind_24_vis_25 = wind_24_vis_25.shape[0]
num_wind_24_vis_25
```

```python
# The mean value of each column for each weather condition?
mean_per_weather_cond = df_renamed.groupby('Weather_Condition')[['Temp_C', 'Dew_Point_Temp_C', 'Rel_Humidity', 'Wind_Speed_km_h', 'Visibility_km']].mean()
mean_per_weather_cond
```

```python
# Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
filtered_criteria = df_renamed[((df_renamed['Weather_Condition'] == 'Clear') & (df_renamed['Rel_Humidity'] > 50)) | (df_renamed['Visibility_km'] > 40)]
filtered_criteria
```

```python
# The number of weather conditions that include snow
snow_in_weather = df_renamed['Weather_Condition'].str.contains('snow', case=False, na=False)
unique_snow_weather = df_renamed[snow_in_weather]['Weather_Condition'].unique()

# Count the number of unique conditions
num_snow_weather = len(unique_snow_weather)

num_snow_weather
```
